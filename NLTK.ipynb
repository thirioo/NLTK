# Stemming

```python
#PRELIMINARY STEPS

import nltk
from nltk.corpus import stopwords
from nltk import FreqDist
```

```python
f = open('input_new.txt','rb')
raw = f.read()
raw = raw.decode("utf-8")
raw = raw.replace('\n',' ')
```

```python
tokens = nltk.word_tokenize(raw)
```

```python
type(tokens)
```

```python
mystopwords = stopwords.words('english')
```

```python
words = [w.lower() for w in tokens if w.isalpha() if w.lower() not in mystopwords]
```

```python
#STEMMING Using PorterStemmer
porter = nltk.PorterStemmer()
```

```python
stem1 = [porter.stem(w) for w in words]
freq1 = FreqDist(stem1)
```

```python
sorted_freq1 = sorted(freq1.items(), key = lambda k:k[1], reverse = True)
```

```python
sorted_freq1
```

```python
freq1.plot(30)
```

```python
#STEMMING Using Lancaster Stemmer
```

```python
Lancaster = nltk.LancasterStemmer()
```

```python
stem2 = [Lancaster.stem(w) for w in words]
```

```python
freq2 = FreqDist(stem2)
```

```python
sorted_freq2 = sorted(freq2.items(), key = lambda k:k[1], reverse = True)
```

```python
sorted_freq2
```

```python
freq2.plot(30)
```

```python
#STEMMING Using WordNet Lemmatizer 
wnl = nltk.WordNetLemmatizer()
nltk.download('wordnet')
```

```python
stem3 = [wnl.lemmatize(w) for w in words]
```

```python
stem3 = [w for w in stem3]
```

```python
freq3 = FreqDist(stem3)
```

```python
sorted_freq3 = sorted(freq3.items(), key = lambda k:k[1], reverse = True)
```

```python
sorted_freq3
```

```python
freq3.plot(30)
```

```python
with open ('stemming_output.txt','w') as f:
    for word, frequency in sorted_freq1:
        f.write(str(word)+'\t'+str(frequency)+'\n')
```
